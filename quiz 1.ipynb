```python
def filter_even_numbers(numbers):
    return list(filter(lambda x: x % 2 == 0, numbers))

numbers = [1, 2, 3, 4, 5, 6, 7, 8]
even_numbers = filter_even_numbers(numbers)
print(even_numbers)
```

```python

import pandas as pd
import numpy as np

data = pd.read_csv('/content/drive/MyDrive/eadg quiz/monthly_revenue.csv')
print(data)
```

```python
import matplotlib.pyplot as plt

data = {
    'Month': ['January', 'February', 'March', 'April', 'May', 'June'],
    'Store A Revenue (in USD)': [12000, 18000, 20000, 22000, 25000, 27000],
    'Store B Revenue (in USD)': [15000, 17000, 16000, 24000, 23000, 28000]
}

df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))

plt.bar(df['Month'], df['Store A Revenue (in USD)'], width=0.4, label='Store A', align='center')

plt.bar(df['Month'], df['Store B Revenue (in USD)'], width=0.4, label='Store B', align='edge')

plt.xlabel('Month')
plt.ylabel('Revenue (in USD)')
plt.title('Monthly Revenue Comparison: Store A vs Store B')

plt.legend()

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

data = {
    'Month': ['January', 'February', 'March', 'April', 'May', 'June'],
    'Store A Revenue (in USD)': [12000, 18000, 20000, 22000, 25000, 27000],
    'Store B Revenue (in USD)': [15000, 17000, 16000, 24000, 23000, 28000]
}

df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))

plt.bar(df['Month'], df['Store A Revenue (in USD)'], width=0.4, label='Store A', align='center', color='skyblue')

plt.bar(df['Month'], df['Store B Revenue (in USD)'], width=0.4, label='Store B', align='edge', color='lightgreen')

plt.title('Monthly Revenue Comparison: Store A vs Store B', fontsize=16, fontweight='bold')
plt.xlabel('Month', fontsize=12)
plt.ylabel('Revenue (in USD)', fontsize=12)

plt.legend(title='Stores', loc='upper left')

plt.xticks(rotation=45)

plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()

plt.show()
```
